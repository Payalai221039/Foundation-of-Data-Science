**Data Representation by Pie Chart**

Pie chart is a type of graph that involves a structural graphic representation of nnumerical proportion.

It can be replaced in most cases by other plots like a bar chart, box plot, dot plot, etc.

It is shown that it is difficult to compare the different section of a given pie chart.

It is also difficult to compare data across different pie charts.

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
plt.axhline(0,color='black',alpha=0.5)
plt.axvline(0,color='black',alpha=0.5)
plt.pie(y)
plt.show()
```

By default the plotting of the first wedge starts from the x-axis and moves counterclockwise:

# Labels

Add labels to the pie chart with the labels parameter.

The labels parameter must be an array with one label for each wedge:

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.axhline(0,color='black',alpha=0.5)
plt.axvline(0,color='black',alpha=0.5)

plt.pie(y, labels = mylabels)
plt.show()
```

# Start Angle

As mentioned the default start angle is at the x-axis, but you can change the start angle by specifying a startangle parameter.

The startangle parameter is defined with an angle in degrees, default angle is 0:

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.axhline(0,color='black',alpha=0.5)
plt.axvline(0,color='black',alpha=0.5)

plt.pie(y, labels = mylabels, startangle = 60)
plt.show()
```

# Explode

Maybe you want one of the wedges to stand out? The explode parameter allows you to do that.

The explode parameter, if specified, and not None, must be an array with one value for each wedge.

Each value represents how far from the center each wedge is displayed:

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0.1, 0]

plt.axhline(0,color='black',alpha=0.5)
plt.axvline(0,color='black',alpha=0.5)

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show()
```

Here **Apple** section is 0.2 unit far away from the center and **Cherries** Section is also 0.1 unit far away from center while **Banana** and **Dates** are on its original position.

# Shadow

Add a shadow to the pie chart by setting the shadows parameter to True:

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.axhline(0,color='black',alpha=0.5)
plt.axvline(0,color='black',alpha=0.5)

plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.show()
```

# Colors

You can set the color of each wedge with the colors parameter.

The colors parameter, if specified, must be an array with one value for each wedge:

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["red", "hotpink", "b", "#4CAF50"]

plt.axhline(0,color='black',alpha=1)
plt.axvline(0,color='black',alpha=1)

plt.pie(y, labels = mylabels, colors = mycolors)
plt.show()
```

You can use Hexadecimal color values, any of the 140 supported color names, or one of these shortcuts:

'r' - Red
'g' - Green
'b' - Blue
'c' - Cyan
'm' - Magenta
'y' - Yellow
'k' - Black
'w' - White

# Legend

To add a list of explanation for each wedge, use the legend() function:

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.axhline(0,color='black',alpha=0.5)
plt.axvline(0,color='black',alpha=0.5)

plt.pie(y, labels = mylabels)
plt.legend(bbox_to_anchor=(1,1))
plt.show()
```

# Legend With Header

To add a header to the legend, add the title parameter to the legend function.

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.axhline(0,color='black',alpha=0.5)
plt.axvline(0,color='black',alpha=0.5)

plt.pie(y, labels = mylabels)
plt.legend(title = "Four Fruits:",bbox_to_anchor=(0,1))
plt.show()
```
